## ORDER to PRODUCT relationship

An Order CONTAINS a Product.

```python
orders.loc[:, ['OrderID', 'ProductID']]
```

Group and Aggregate to identify the number of products in each order.

```python
orders.groupby(['OrderID']).\
aggregate(num_rows = ('OrderID', 'size'),
          num_products = ('ProductID', 'nunique')).\
reset_index()
```

One row is a product in an order.

```python
orders.groupby(['OrderID', 'ProductID']).size().reset_index(name='num_rows')
```

```python
orders.groupby(['OrderID', 'ProductID']).size().reset_index(name='num_rows').\
num_rows.value_counts()
```

We will also bring in two other properties.

```python
orders.loc[:, ['OrderID', 'ProductID', 'Quantity', 'UnitPrice']]
```

## Employee to Order Relatioship

An Employee SOLD an Order.

```python
orders.loc[:, ['OrderID', 'ProductID', 'EmployeeID']]
```

```python
orders.groupby(['OrderID', 'EmployeeID']).\
aggregate(num_rows = ('OrderID', 'size'),
          num_products = ('ProductID', 'nunique')).\
reset_index()
```
